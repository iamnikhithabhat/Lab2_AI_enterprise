# Below is the process of Model Training

```python
x = df_cancer.drop(['target'], axis=1)
y= df_cancer['target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# We perform normalization process to ensure all the values are between 0 and 1 for better accuracy 

train_minimum = X_train.min()
train_range = (X_train - train_minimum).max()
X_train_scaled = (X_train - train_minimum)/ train_range
```

```python
sns.scatterplot(x = X_train['mean compactness'], y = X_train['mean concavity'], hue = y_train) #before normalisation
```

```python
sns.scatterplot(x = X_train_scaled['mean compactness'], y = X_train_scaled['mean concavity'], hue = y_train) #after normalisation
```

```python
#similarly we normalise test values
test_minimum = X_test.min()
test_range = (X_test - test_minimum).max()
X_test_scaled = (X_test - test_minimum)/ test_range
```

```python
#create a Naive Bayes model and train the model
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(X_train_scaled, y_train)
new_nb=naive_bayes.predict(X_test_scaled)
new_nb
```

```python
from sklearn.metrics import accuracy_score
print("Naive Bayes score: ",naive_bayes.score(X_test_scaled, y_test))
```

```python
accuracy_score(y_test,new_nb)
```

accuracy_score and .score are used for same purpose

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,new_nb)
```

There are 7 values that are predicted wrongly.Out of which 5 are false positives and 2 are false negatives.
